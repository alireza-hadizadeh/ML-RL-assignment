```python
import numpy as np
import matplotlib.pyplot as plt

# Data (hours of study and test score)
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])

# Step 1: Means
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Step 2: Calculate slope (a) and y-intercept (b)
numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)
a = numerator / denominator
b = mean_y - a * mean_x

print(f"Line equation: Y = {a:.2f}X + {b:.2f}")

# Step 3: Plot data and regression line
plt.scatter(X, Y, color='blue', label='Data')
regression_line = a * X + b
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.xlabel('Hours of Study (X)')
plt.ylabel('Test Score (Y)')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Prediction
x_new = 6
y_pred = a * x_new + b
print(f"If you study for {x_new} hours, your predicted score is: {y_pred:.2f}")
```
