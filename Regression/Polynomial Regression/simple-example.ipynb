```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])
n = len(X)

# Step 2: Compute the required powers and products
X2 = X**2
X3 = X**3
X4 = X**4
XY = X * Y
X2Y = X2 * Y

# Step 3: Calculate the sums needed for the system of equations
sum_X = np.sum(X)
sum_X2 = np.sum(X2)
sum_X3 = np.sum(X3)
sum_X4 = np.sum(X4)
sum_Y = np.sum(Y)
sum_XY = np.sum(XY)
sum_X2Y = np.sum(X2Y)

# Step 4: Create the matrix A and vector B
A = np.array([
    [sum_X4, sum_X3, sum_X2],
    [sum_X3, sum_X2, sum_X],
    [sum_X2, sum_X, n]
])
B = np.array([sum_X2Y, sum_XY, sum_Y])

# Step 5: Solve the system A * [a, b, c] = B
coeffs = np.linalg.solve(A, B)
a, b, c = coeffs
print(f"The polynomial equation is: Y = {a:.2f}X² + {b:.2f}X + {c:.2f}")

# Step 6: Plot the data and the fitted curve
x_range = np.linspace(min(X), max(X), 100)
y_range = a * x_range**2 + b * x_range + c

plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(x_range, y_range, color='green', label='Fitted Polynomial Curve (deg 2)')
plt.xlabel('Study Hours (X)')
plt.ylabel('Exam Score (Y)')
plt.title('Polynomial Regression (Degree 2)')
plt.grid(True)
plt.legend()
plt.show()

# Step 7: Prediction example
x_new = 6
y_pred = a * x_new**2 + b * x_new + c
print(f"Predicted Y for X = {x_new} is: {y_pred:.2f}")
```
